```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('netflix_titles.csv')
df = df.drop(df.columns[0], axis=1)
```

```python
df.head()
```

```python
eixo_x = df['listed_in'].value_counts().keys()
eixo_y = df['listed_in'].value_counts().values
```

```python
plt.bar(eixo_x, eixo_y)
plt.show
```

```python
plt.bar(eixo_x, eixo_y)
plt.title('Filmes por categoria')
plt.xlabel('Gênero de Filme')
plt.ylabel('Quantidade de filmes')
plt.xticks(rotation=45)
plt.show
```

```python
## Analisando a quantidade de filmes por ano de lançamento

eixo_x = df['release_year'].value_counts().sort_index().keys()
eixo_y = df['release_year'].value_counts().sort_index().values
```

```python
plt.plot(eixo_x, eixo_y)
plt.title('Lançamento de filmes por ano')
plt.xlabel('Ano')
plt.ylabel('Quantidade de filmes')
plt.xticks(rotation=45)
```
